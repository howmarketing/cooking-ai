```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the markers
# pink_markers =  [[1, 2], [2, 1]]
# green_markers = [[4, 3], [3, 5]]

def plot_axis_graph(pink_markers, green_markers):

    # Define the line of slope -1
    x_vals = np.linspace(0, 7, 400)
    y_vals = -x_vals + 7

    # print("x_vals: ", x_vals)
    # print("y_vals: ", y_vals)

    # Create the plot
    plt.figure(figsize=(5, 5))
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.2)

    # Plot the markers
    plt.scatter(pink_markers[0], pink_markers[1], color='pink', label='Losers', s=80, edgecolor='gray')
    plt.scatter(green_markers[0], green_markers[1], color='green', label='Winners', s=80, edgecolor='gray')

    # Plot the line of slope -1
    plt.plot(x_vals, y_vals, color='blue', linestyle='--', label='Line with slope -1')

    # Set the axis limits
    plt.xlim(0, 8)
    plt.ylim(0, 8)

    # Add labels and legend
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Axis Graph with Line of Slope -1 Dividing Markers')
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def create_markers(markers = [[1,2], [2,1]]):
    x = []
    y = []
    for marker in markers:
        x.append(marker[0])
        y.append(marker[1])
    return [x, y]

markers = {
    "green": [],
    "pink": [],
}

markers["pink"] = create_markers([[4,2], [3,2]])
markers["green"] = create_markers([[4,3], [3,5]])

# plot_axis_graph(markers["pink"], markers["green"])

print("pink", markers["pink"])
print("green", markers["green"])
```

```python
markers["pink"] = create_markers([[2,2], [3,1]])
markers["green"] = create_markers([[3,5], [5,2]])
plot_axis_graph(markers["pink"], markers["green"])
```
